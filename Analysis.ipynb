### 3. Is there a correlation between the amount of RAM and the price of the laptop?

```python
# Convert RAM to numeric value
df['ram_gb'] = df['ram_gb'].str.replace(' GB', '').astype(int)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ram_gb', y='Price')
plt.title('RAM vs Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.show()

# Calculate the correlation
ram_price_correlation = df['ram_gb'].corr(df['Price'])
ram_price_correlation
```

### 4. What are the most common processor brands and their average prices?

```python
# Calculate the processor brand counts
processor_brand_counts = df['processor_brand'].value_counts()

# Calculate the average price by processor brand
average_prices_by_processor_brand = df.groupby('processor_brand')['Price'].mean()

# Plot processor brand counts
plt.figure(figsize=(14, 7))
sns.barplot(x=processor_brand_counts.index, y=processor_brand_counts.values, palette='viridis')
plt.title('Processor Brand Counts')
plt.xlabel('Processor Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot average prices by processor brand
plt.figure(figsize=(14, 7))
sns.barplot(x=average_prices_by_processor_brand.index, y=average_prices_by_processor_brand.values, palette='viridis')
plt.title('Average Price by Processor Brand')
plt.xlabel('Processor Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

### 5. How does the presence of a touchscreen affect the price?

```python
average_price_touchscreen = df.groupby('Touchscreen')['Price'].mean()
average_price_touchscreen
```

```python
# Prepare data for pie chart
labels = average_price_touchscreen.index
sizes = average_price_touchscreen.values

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Average Price of Laptops With and Without Touchscreen')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### 6. What is the distribution of SSD sizes and how do they affect laptop prices?

```python
# Convert all values in 'ssd' column to strings, then replace ' GB' and convert to integers
df['ssd'] = df['ssd'].astype(str).str.replace(' GB', '').astype(int)

# Plot the distribution of SSD sizes
plt.figure(figsize=(10, 6))
sns.histplot(df['ssd'], bins=20, kde=True)
plt.title('Distribution of SSD Sizes')
plt.xlabel('SSD Size (GB)')
plt.ylabel('Count')
plt.show()
```

### 7. What is the average price of laptops based on the type of operating system?

```python
# Calculate the average price by OS
average_price_by_os = df.groupby('os')['Price'].mean()

# Plot the average price by OS
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_os.index, y=average_price_by_os.values, palette='viridis')
plt.title('Average Price of Laptops by Operating System')
plt.xlabel('Operating System')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```
